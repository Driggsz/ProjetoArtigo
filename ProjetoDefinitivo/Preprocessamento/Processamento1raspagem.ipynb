```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
dados = pd.read_csv('brasileirao_serie_a.csv')
```

```
dados.count()
```

#filtrando dados <= 2005

```
# Filtrar por data
dados_filtrado = dados[dados['ano_campeonato'] >= 2006]

for coluna in dados_filtrado.columns:
  # Contar valores preenchidos e nulos
  preenchidos = dados_filtrado[coluna].notna().sum()
  nulos = dados_filtrado[coluna].isna().sum()
  total = preenchidos + nulos

  # Calcular porcentagens
  porcentagem_preenchidos = (preenchidos / total) * 100
  porcentagem_nulos = (nulos / total) * 100

  # Apresentar resultados
  print(f"Coluna: {coluna}")
  print(f"Preenchidos: {preenchidos} ({porcentagem_preenchidos:.2f}%)")
  print(f"Nulos: {nulos} ({porcentagem_nulos:.2f}%)")
  print("-" * 50)
```

#Dropando colunas e linhas < 2006

```
#dropando colunas
dados = dados.drop(['data',
              'publico_max',
              'escanteios_mandante',
              'escanteios_visitante',
              'faltas_mandante',
              'faltas_visitante',
              'chutes_bola_parada_mandante',
              'chutes_bola_parada_visitante',
              'defesas_mandante',
              'defesas_visitante',
              'impedimentos_mandante',
              'impedimentos_visitante',
              'chutes_mandante',
              'chutes_visitante',
              'chutes_fora_mandante',
              'chutes_fora_visitante'], axis=1)
```

```
dados = dados.drop(dados[dados['ano_campeonato'] < 2006].index)
```

#Calculando a media das colunas e preenchendo onde está vazio

```
media = dados['publico'].mean()
dados['publico'] = dados['publico'].fillna(media)
```

```
media = dados['colocacao_mandante'].mean()
dados['colocacao_mandante'] = dados['colocacao_mandante'].fillna(media)
```

```
media = dados['colocacao_visitante'].mean()
dados['colocacao_visitante'] = dados['colocacao_visitante'].fillna(media)
```

```
media = dados['valor_equipe_titular_mandante'].mean()
dados['valor_equipe_titular_mandante'] = dados['valor_equipe_titular_mandante'].fillna(media)
```

```
media = dados['valor_equipe_titular_visitante'].mean()
dados['valor_equipe_titular_visitante'] = dados['valor_equipe_titular_visitante'].fillna(media)
```

```
media = dados['idade_media_titular_mandante'].mean()
dados['idade_media_titular_mandante'] = dados['idade_media_titular_mandante'].fillna(media)
```

```
media = dados['idade_media_titular_mandante'].mean()
dados['idade_media_titular_mandante'] = dados['idade_media_titular_mandante'].fillna(media)
```

```
media = dados['gols_mandante'].mean()
dados['gols_mandante'] = dados['gols_mandante'].fillna(media)
```

```
media = dados['gols_visitante'].mean()
dados['gols_visitante'] = dados['gols_visitante'].fillna(media)
```

```
media = dados['gols_1_tempo_mandante'].mean()
dados['gols_1_tempo_mandante'] = dados['gols_1_tempo_mandante'].fillna(media)
```

```
media = dados['gols_1_tempo_visitante'].mean()
dados['gols_1_tempo_visitante'] = dados['gols_1_tempo_visitante'].fillna(media)
```

```
#Dropando o restante dos dados que tem nulos
dados = dados.dropna()
```
